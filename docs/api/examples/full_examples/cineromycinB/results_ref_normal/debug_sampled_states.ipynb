```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.load("traj_lambda0.00.npz", allow_pickle=True)["arr_0"].item()
y = np.load("traj_lambda0.50.npz", allow_pickle=True)["arr_0"].item()
sx = x["state_trace"]  
sy = y["state_trace"]  
```

```python
#print(sx[::100])
```

```python
%matplotlib inline
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.hist(sx, bins=100)
fig.tight_layout()
fig.show()
```

```python
%matplotlib inline
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.hist(sy, bins=100)
fig.tight_layout()
fig.show()
```
