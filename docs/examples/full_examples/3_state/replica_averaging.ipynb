<h5 style="text-align:center;font-size: 18pt;font-weight:bold">Introduction</h5>


<!--Why add replica averaging? What does it serve?-->

Replicas are several copies of the model. Using multiple replicas is meant to represent a small version of the structural ensemble that is present during the experimental measurements.  Ultimately, the goal is to increase the resolution of structural information that often gets “blurred out” due to motional averaging.

**By applying replica-averaging, we can overcome challenges like integration of various information from different experiments and prior knowledge of the system, properly deals with errors and resolve ensemble-averaged data.**



<h5 style="text-align:center;font-size: 18pt;font-weight:bold">Theory</h5>

\begin{equation}
P(X,\sigma | D) \propto Q(D|X,\sigma) P(X) P(\sigma)
\end{equation}

\begin{equation}
Q(D | X, \sigma)=\prod_{j} \frac{1}{\sqrt{2 \pi \sigma^{2}}} \exp \left(-\left[d_{j}(X)-d_{j}^{\exp }\right]^{2} / 2 \sigma^{2}\right)
\end{equation}


<h5 style="align:left;font-size: 18pt;font-weight:bold">Gaussian noise model</h5>

$\sigma^{B}$ is the Bayesian error sampled by Monte Carlo

$\sigma^{SEM}$ is the standard error of the mean of the forward model




<!--
\begin{equation}
Q(D | \tilde{f}, \sigma^{B})= \prod^{N_{d}}_{j} \frac{1}{\sqrt{2 \pi (\sigma^{B})^{2}}} \exp \left(-\left[d_{j} - f_{j}(X)\right]^{2} / 2 (\sigma^{B})^{2}\right),
\end{equation}
-->

\begin{equation}
P(\tilde{f} | X, \sigma^{SEM})= \frac{1}{\sqrt{2 \pi (\sigma^{SEM})^{2}}} \exp \left(-\left[\tilde{f} - f(X)\right]^{2} / 2 (\sigma^{SEM})^{2}\right),
\end{equation}

where $f(x)$ is the forward model averaged across replicas given by

$$f(x) = \frac{1}{N}\sum^{N}_{r=1} f(x_{r})$$

&\propto \prod^{N}_{r} Q(D|\tilde{f_{r}},\sigma^{B}_{r}) P(\tilde{f_{r}}| X,\sigma^{SEM}_{r}) P(X_{r}) P(\sigma^{SEM}_{r}) P(\sigma^{B}_{r})\\

-->


$$ \prod^{N}_{r} P(X_{r}) \prod^{N_{d}}_{j} \frac{1}{\sqrt{2 \pi (\sigma^{B}_{r})^{2}}} \exp \left(-\left[d_{j} - f_{j}(X)\right]^{2} / 2 (\sigma^{B}_{r})^{2}\right)  P(\sigma^{SEM}_{r}) P(\sigma^{B}_{r}) $$


\begin{align}
P(X,\tilde{f},\sigma^{SEM},\sigma^{B} | D) &\propto Q(D|\tilde{f},\sigma^{B}) P(\tilde{f}| X,\sigma^{SEM}) P(X) P(\sigma^{SEM}) P(\sigma^{B})\\
&\propto \prod^{N}_{r} Q(D|\tilde{f_{r}},\sigma^{B}_{r}) P(\tilde{f_{r}}| X,\sigma^{SEM}_{r}) P(X_{r}) P(\sigma^{SEM}_{r}) P(\sigma^{B}_{r})\\
\end{align}

\begin{equation}
P(X,\tilde{f},\sigma | D)\propto \prod^{N}_{r} P(X_{r}) \prod^{N_{d}}_{j} \frac{1}{\sqrt{2 \pi (\sigma_{r,i})^{2}}} \exp \left(-\left[d_{j} - f_{j}(X)\right]^{2} / 2 (\sigma_{r,i})^{2}\right)  P(\sigma_{r,i}),
\end{equation}


where $\sigma_{r} = \sqrt{(\sigma^{SEM}_{r})^{2} + (\sigma^{B}_{r})^{2}}$




**Objective function or Loss function (in our case):** Minimizing the sum of squared errors

For $\sigma^{B}_{r}$

\begin{equation}
\chi^{2}(X)=\sum_{j} w_{j}\left(d_{j}(X)-d_{j}^{\exp }\right)^{2}
\end{equation}

For $\sigma^{SEM}_{r}$

\begin{equation}
\chi^{2}(X)=\sum_{j} w_{j}\left(d_{j} - f_{j}(X)\right)^{2}
\end{equation}
where $w_j$ is a weight parameter that is designed for equivalent observables.




The error parameters $\{ \sigma_{r,i}\}$ determine the intensity of the data-restraints and are defined as $\sigma^{2}_{r,i} = (\sigma^{B}_{r,i})^{2} + (\sigma^{SEM}_{r,i})^{2}$, where $\sigma^{B}_{r,i}$ is the Bayesian
error sampled by MC, and $\sigma^{SEM}_{r,i}$ the standard error of the mean.



\begin{equation}
-\ln P(X, \sigma | D)=\left(N_{j}+1\right) \ln \sigma+\chi^{2}(X) / 2 \sigma^{2}-\ln Q_{\text{ref}}+\left(N_{j} / 2\right) \ln 2 \pi  -\ln P(X)
\end{equation}

````python
    def compute_logZ(self):
        """Compute reference state logZ for the free energies to normalize."""

        Z = 0.0
        for s in self.ensemble:
            Z +=  np.exp( -np.array(s[0].energy, dtype=np.float128) )
        self.logZ = np.log(Z)


    def neglogP(self, states, parameters, parameter_indices):
        """Return -ln P of the current configuration.

        Args:
            state(list): the new conformational state from Sample()
            parameters(list): a list of the new parameters for each of the restraints
            parameter_indices(list): parameter indices that correspond to each restraint
        """

        result = 0
        sse = self.get_SSE(states)
        for state in states:
            s = self.ensemble[int(state)] # Current Structure (list of restraints)
            result += s[0].energy + self.logZ  # Grab the free energy of the state and normalize
            for i,R in enumerate(s):
                result += R.compute_neglogP(parameters[i], parameter_indices[i], sse[i]+s[i].sse)
        return result

````
