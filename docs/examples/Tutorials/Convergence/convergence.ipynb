<h1 style="align: center;font-size: 18pt;">Convergence</h1>
<hr style="height:2.5px">

This tutorial shows the user how to use the **convergence submodule** within `biceps` to determine if the MCMC sampling has converged.  

<hr style="height:2.5px">

```python
import biceps
```

Loading a `biceps` MCMC trajectory file of 1M steps...

```python
%matplotlib inline
C = biceps.Convergence(
    trajfile='../MP_Lambdas/results/traj_lambda0.00.npz',
    resultdir="./")
```

```python
C.plot_traces(fname="traces.pdf", xlim=(0, 1e6))
```

<h1 style="align: center;font-size: 18pt;">Computing the autocorrelation (<a href="#Second-method:-using-autocorrelation-function-(preferred),-where"><u>second method is preferred</u></a>)</h1>

<h1 style="text-align: left;font-size: 18pt;">First method: using a single exponential fitting</h1>



```python
C.get_autocorrelation_curves(method="exp", maxtau=5000)
```

<h1 style="text-align: left;font-size: 18pt;">Using $\tau_{auto}$ <a href="#First-method:-using-a-single-exponential-fitting">from exponential fitting</a> for processing (JSD distributions)</h1>



```python
C.process(nblock=5, nfold=10, nround=100, savefile=False,
    plot=False, normalize=True)
```

<h1 style="text-align: left;font-size: 18pt;">Second method: using autocorrelation function (preferred), where</h1>

$$\tau_{auto} = \int C_{\tau} d\tau$$

```python
C.get_autocorrelation_curves(method="auto", maxtau=1000)
```

<h1 style="text-align: left;font-size: 18pt;">Using $C_{\tau}$ and $\tau_{auto}$ from autocorrelation function for processing (output block averaging and JSD distributions)</h1> 

```python
C.process(nblock=5, nfold=10, nround=100, savefile=False,
    plot=True, block=True, normalize=True)
```
